# **Cost - Revenue Relation**

#### In this project we will performing analysis on previously collected data based on production budget and worldwide gross revenue of multiple movies from which we will be getting relation between both the variables and get information and insight on how much successful a movie is as compared to the production budget.

We import pandas, matplotlib and LinearRegression from python library for using it further for reading csv, visualising data and for regression. 

```python
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
```

We now read the csv file containing the data on which we will be performing analysis and get the required insights.

```python
data = pd.read_csv('cost_revenue.csv')
```

```python
data.describe()
```

We now define two variables X and y which will be used to visualise our data in a graph.

```python
X = DataFrame(data, columns = ['production_budget_usd'])
y = DataFrame(data, columns = ['worldwide_gross_usd'])
```

We now will use a scatter graph to represent our data. We will provide the chart with a title and label the x and y axes, and update the size of the chart according to our convinience 

```python
plt.figure(figsize = (10,6))
plt.scatter(X,y, alpha = 0.3)
plt.title('Film Cost vs Global Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('Worldwide gross $')
plt.ylim(0,3000000000)
plt.xlim(0,450000000)
plt.show()
```

#### Regression

```python
regression = LinearRegression()
regression.fit(X,y)
```

Slope coefficient

```python
regression.coef_
```

Intercept

```python
regression.intercept_
```

Now using linear regression we will predict the value for each budget value and plot it on the graph. This will help us clarify the relation between the variables.

```python
plt.figure(figsize = (10,6))
plt.scatter(X,y, alpha = 0.3)
plt.plot(X, regression.predict(X), color = 'red', linewidth = 4)
plt.title('Film Cost vs Global Revenue')
plt.xlabel('Production Budget $')
plt.ylabel('Worldwide gross $')
plt.ylim(0,3000000000)
plt.xlim(0,450000000)
plt.show()
```

Goodness of fit
- This will explain the amount of variation in film revenue explained by the film's budget.

```python
regression.score(X,y)
```

## Therefore this analysis explains that the film's success does not depend solely on the production budget, it only affects it approximately 55%, therefore we cannot predict a movie's success based only on it's budget.
